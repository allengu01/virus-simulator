```python
import math
import random
import matplotlib.pyplot as plt
%matplotlib inline

def rand_r(max_r):
    return max_r*math.sqrt(random.random())

def rand_angle(max_angle):
    return max_angle*random.random()
```

```python
polar_coords = []
n=1000
mr=10 # max radius
for i in range(n):
    polar_coords.append((rand_r(mr), rand_angle(2*math.pi)))
```

```python
rect_coordsX = []
rect_coordsY = []
for j in polar_coords:
    x=j[0]*math.cos(j[1])
    y=j[0]*math.sin(j[1])
    rect_coordsX.append(x)
    rect_coordsY.append(y)
```

```python
plt.figure("Random Generated Circle", figsize=(10,10))
plt.title("Points uniformly and randomly generated in circle")
plt.xlabel("X")
plt.ylabel("Y")
plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.scatter(rect_coordsX, rect_coordsY)
plt.show()
```
